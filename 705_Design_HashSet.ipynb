```python
#================ 705. Design HashSet ==================
class MyHashSet:
  def __init__(self):
    self.hash_set = [False]*100001
  
  def add(self, key):
    self.hash_set[key] = True
  
  def remove(self, key):
    self.hash_set[key] = False
  
  def contains(self, key):
    return self.hash_set[key]

hash = MyHashSet()
hash.add(1);         
hash.add(2);         
hash.contains(1);   
print(hash.hash_set) 
hash.remove(2);          
print(hash.contains(2)); 
```
